```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph()
G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edge(2, 3)
G.add_edge(3, 4)
G.add_edge('a', 'b')
G.add_edge('a', 'c')
G.add_edge('b', 'c')
```

```python
nx.draw_networkx(G)
```

```python
list(nx.weakly_connected_components(G))
```

```python
nx.is_directed_acyclic_graph(G)
```

```python
nodes = next(filter(lambda s: "a" in s, nx.weakly_connected_components(G)))
nx.draw_networkx(G.subgraph(nodes))
```
