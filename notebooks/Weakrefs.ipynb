```python
from weakref import ref
```

```python

valid = True


def invalidate(ref):
    print(f"Invalidating ref")
    global valid
    valid = False


class BigClass:
    def hello(self):
        print("Hello")
```

```python

a = BigClass()

r = ref(a, invalidate)

assert valid

assert r() == a

r().hello()

del a

assert not valid

assert r() is None
```

```python
from weakref import proxy

valid = True

a = BigClass()
p = proxy(a, invalidate)

p.hello()
```

```python
del a
```

```python
p.hello()
```

Hashability
=====

Can I make proxy objects hashable?

```python
a = BigClass()
p = proxy(a, invalidate)
```

```python
hash(p)
```

```python
def myhash(self, *args, **kw):
    return 123321

p.__hash__ = myhash
hash(p)
```

```python
from weakref import weakproxy
```

```python
a = BigClass()
b = BigClass()
p_a = proxy(a, invalidate)
p_b = proxy(b, invalidate)
```

```python
{p_a, p_b}
```
